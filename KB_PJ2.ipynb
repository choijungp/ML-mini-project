```python
from pycaret.classification import *
```

```python
data.info()
```

```python
categorical_features = ['이름','팀명','경기날짜', '포지션', '경기번호']
```

```python
clf = setup(data = data, target = '결과',
            session_id = SEED,    
            categorical_features=categorical_features,
            categorical_imputation='mode',
            train_size=0.8,
            profile = True # 분석을 끝내고 보고서 생성, False : 보고서 생성 안함
            )
```

```python
best_models = compare_models(sort='Accuracy',n_select=3)
```

```python
blended = blend_models(estimator_list=best_models, fold=5)
```

```python
mymodels = tune_model(blended)
```

```python
final_model = finalize_model(mymodels)
```

```python
test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Dataset/auto_validation.csv', index_col=0)
```

```python
test.head()
```

```python
pred = predict_model(final_model, data=test)
pred
```

```python
prediction = pred['Label']
```

```python
prediction
```

```python
from sklearn.metrics import confusion_matrix

cm = pd.DataFrame(confusion_matrix(pred['결과'], prediction),columns = pred['결과'].unique(),
                  index = pred['결과'].unique())
cm
```
